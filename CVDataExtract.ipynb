```python
###########################################################################################################
#
#Path of CV files
PATH = "C:\\Users\\Vasu\\Downloads\\React.JS Naukri MM-20211214T070121Z-001\\React.JS Naukri MM"


###########################################################################################################
import docx2txt
import PyPDF2
import re
import glob
import pandas as pd
import pdfplumber
Name = []
Email = []
Number = []
pdfPath = PATH+"\\*.pdf"
print(pdfPath)
for file in glob.glob(pdfPath):
#     pdffileobj=open(file,'rb')
#     pdfreader=PyPDF2.PdfFileReader(pdffileobj) 
#     x=pdfreader.numPages
#     pageobj = pdfreader.getPage(x-1)
#     my_text=pageobj.extractText()
    x = []
    y = []
    Name.append(file.split('\\')[-1].replace(".pdf",""))
    try:
        with pdfplumber.open(file) as pdf:
            first = pdf.pages[0]
            my_text = first.extract_text()
        pattern_email = re.compile(r'[a-zA-Z0-9-\.]+@[a-zA-Z-\.]*\.(com|edu|net)')
        matches_email = pattern_email.finditer(my_text)
        pattern_number = re.compile("\d{3}\d{3}\d{4}")
        matches_number = pattern_number.finditer(my_text)
        for match in matches_email:
            x.append(match.group(0))
            print(match.group(0))
        Email.append(x)
        for match in matches_number:
            y.append(match.group(0))
            print(match.group(0))
        Number.append(y)
    except:
        Email.append(" ")
        Number.append(" ")
        
docxPath = PATH+"\\*.docx"
for file in glob.glob(docxPath):
    x = []
    y = []
    my_text = docx2txt.process(file)
    pattern_email = re.compile(r'[a-zA-Z0-9-\.]+@[a-zA-Z-\.]*\.(com|edu|net)')
    matches_email = pattern_email.finditer(my_text)
    pattern_number = re.compile("\d{3}\d{3}\d{4}")
    matches_number = pattern_number.finditer(my_text)
    Name.append(file.split('\\')[-1].replace(".docx",""))
    for match in matches_email:
        x.append(match.group(0))
        print(match.group(0))
    Email.append(x)
    for match in matches_number:
        y.append(match.group(0))
        print(match.group(0))
    Number.append(y)
Data = pd.DataFrame({"Name":Name,"Email":Email,"Number":Number})
save_path = PATH+"\\CV_Data.csv"
Data.to_csv(save_path,index=False)


# Data stored in the same folder where CVs are.
```
